```python
from google_play_scraper import Sort, reviews_all

us_reviews = reviews_all(
    'com.rhbgroup.rhbmobilebanking',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
df_rhb_google = pd.DataFrame(np.array(us_reviews),columns=['review'])

df_rhb_google = df_rhb_google.join(pd.DataFrame(df_rhb_google.pop('review').tolist()))

df_rhb_google.tail()
```

```python
df_rhb_google.to_csv('rhb_google_review.csv')
```
